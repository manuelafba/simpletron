```python
# Instruções
READ = 10 # Lê uma palavra do teclado para uma posição na memória
WRITE = 11 # Escreve uma palavra de uma determinada posição na memória
LOAD = 20 # Carrega uma palavra de uma posição específica
STORE = 21 # Armazena uma palavra de uma posição específica
ADD = 30 # Soma uma palavra de uma posição com a que está no acumulador (o resultado permanece no ACC)
SUBTRACT = 31 # Subtrai uma palavra de uma posição com a que está no acumulador
DIVIDE = 32 # Divide uma palavra de uma posição com a que está no acumulador
MULTIPLY = 33 # Soma uma palavra de uma posição com a que está no acumulador
BRANCH = 40 # Jumper incondicional, desvia para uma posição específica na memória
BRANCHNEG = 41 # Desvia para uma posição específica na memória se o valor no ACC for negativo
BRANCHZERO = 42 # Desvia para uma posição específica na memória se o valor no ACC for igual a zero
HALT = 43 # Parada do programa

mp = [0]*100 # Lista que representa a memória principal
inst = 0 # Instrução
end = 0 # Endereço
acc = 0 # Acumulador
i = 0 # Variável para incrementar a posição na MP

print("===        Bem vindo ao Simpletron no Jupyter!        ===")
print("=== Por favor, digite seu programa, uma instrução     ===")
print("=== (ou palavra de dados) de cada vez. Eu digitarei o ===")
print("=== número da posição e um ponto de interrogação (?). ===")
print("=== Então você digita a palavra para aquela posição.  ===")
print("=== Digite a sentinela -99999 para terminar a entrada ===")
print("=== de seu programa                                   ===")  

while inst != -99999: # Enquanto não for digitado a sentinela -99999
    inst = int(input(f"{end}?: "))
    if inst == -99999: 
        print("== Carga do programa completa       ==")
        print("== Iniciando a execução do programa ==")
    else:
        mp.insert(end, inst) # Insere o endereço e a instrução na MP
    end += 1 # Incrementa o endereço a cada instrução digitada

while i != 43: # Enquanto a instrução não for o HALT
    codOp = int(mp[i])//100 # Primeiros dois dígitos para representar o código de operação
    operando = int(mp[i]) % 100 # Últimos dois dígitos para representar o operando

    if codOp == READ:
        mp[operando] = int(input(f'Digite um valor para a {operando}ª célula: '))
    if codOp == WRITE:
        print(mp[operando]) 
    if codOp == LOAD:
        acc = mp[operando]
    if codOp == STORE:
        mp[operando] = acc
    if codOp == ADD:
        acc += mp[operando] # ACC <- ACC + Op.
    if codOp == SUBTRACT:
        acc -= mp[operando]
    if codOp == DIVIDE:
        if acc != 0 and mp[operando] != 0:
            acc /= mp[operando]
        else:
            print("ERRO! Divisão por zero!") # A mensagem de erro irá aparecer no caso de uma divisão por zero, ou seja, quando o ACC for zero
    if codOp == MULTIPLY:
        acc *= mp[operando]
    if codOp == BRANCH:
        if 0 <= operando <= 99: # Jump incondicional
            i = operando # Atualiza a posição, que será o endereço indicado pelo operando
            continue
    if codOp == BRANCHZERO:
        if 0 <= operando <= 99 and acc == 0: # Jump se o valor do R0 for 0
            i = operando
            continue
    if codOp == BRANCHNEG:
        if 0 <= operando <= 99 and acc < 0: # Jump se o valor do R0 for negativo
            i = operando
            continue
    if codOp == HALT:
        break # Quebra do loop

    i += 1

print("=== O programa terminou de forma normal ===")
print('')
print("REGISTRADORES")
print(f"Acumulador: {acc}")
print(f"Counter: {i}")
print(f"RI: {inst}")
print(f"Código de operação: {codOp}")
print(f"Operando: {operando}")
print('')

print("==== MEMÓRIA ====")
for i in range(len(mp)): # Percorre a memória do primeiro endereço até o último, obtido pelo "len(memory)"
    endereco = f"{i:02d}" # Endereço formatado para dois dígitos
    print(endereco, "-", mp[i]) 
```
