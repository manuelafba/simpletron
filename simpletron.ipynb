```python
# Instruções
READ = 10
WRITE = 11
LOAD = 20
STORE = 21
ADD = 30
SUBTRACT = 31
DIVIDE = 32
MULTIPLY = 33
BRANCH = 40
BRANCHNEG = 41
BRANCHZERO = 42
HALT = 43

mp = [0]*100
inst = 0
end = 0
r0 = 0
i = 0

print("=== Por favor, digite seu programa, uma instrução     ===")
print("=== (ou palavra de dados) de cada vez. Eu digitarei o ===")
print("=== número da posição e um ponto de interrogação (?). ===")
print("=== Então você digita a palavra para aquela posição   ===")
print("=== Digite a sentinela -99999 para terminar a entrada ===")
print("=== de seu programa                                   ===")  

while inst != -99999: # Enquanto não for digitado a sentinela -99999
    inst = int(input(f"{end}?: "))
    if inst == -99999:
        print("== Carga do programa completa       ==")
        print("== Iniciando a execução do programa ==")
    else:
        mp.insert(end, inst) # Insere o endereço e a instrução na MP
    end += 1

while i != 43: # Enquanto a instrução não for o HALT
    codOp = int(mp[i])//100 # Primeiros dois dígitos: código de operação
    operando = int(mp[i]) % 100 # Últimos dois dígitos: operando

    if codOp == READ:
        mp[operando] = int(input(f'Digite um valor para a {operando}ª célula: '))
    if codOp == WRITE:
        print(mp[operando]) 
    if codOp == LOAD:
        r0 = mp[operando]
    if codOp == STORE:
        mp[operando] = r0
    if codOp == ADD:
        r0 += mp[operando]
    if codOp == SUBTRACT:
        r0 -= mp[operando]
    if codOp == DIVIDE:
        if r0 != 0 and mp[operando] != 0:
            r0 /= mp[operando]
        else:
            print("ERRO! Divisão por zero!")
    if codOp == MULTIPLY:
        r0 *= mp[operando]
    if codOp == BRANCH:
        if 0 <= operando <= 99: # Jump incondicional
            i = operando
            continue
    if codOp == BRANCHZERO:
        if 0 <= operando <= 99 and r0 == 0: # Jump se o valor do R0 for 0
            i = operando
            continue
    if codOp == BRANCHNEG:
        if 0 <= operando <= 99 and r0 < 0: # Jump se o valor do R0 for negativo
            i = operando
            continue
    if codOp == HALT:
        break

    i += 1

print("=== O programa terminou de forma normal ===")
print('')
print("REGISTRADORES")
print(f"R0: {r0}")
print(f"Counter: {i}")
print(f"RI: {inst}")
print(f"Código de operação: {codOp}")
print(f"Operando: {operando}")
print('')

print("==== MEMÓRIA ====")
print("Célula | Conteúdo")
print("-----------------")
for i in range(100):
    endereco = f"{i:02d}" #front end
    cont = str(mp[i])
    print(f"{endereco}    | {cont}")
```
